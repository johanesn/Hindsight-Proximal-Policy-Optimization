```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

# Load Data

```python
import pickle

# HPPO Clipped Normal Reward Normalized
with open('./Test Reward Plot/test_rewards_normal', 'rb') as fp1:
    test_rewards     = pickle.load(fp1)
with open('./Loss Plot/mean_actor_loss_normal', 'rb') as fp2:
    mean_actor_loss  = pickle.load(fp2)
with open('./Loss Plot/mean_critic_loss_normal', 'rb') as fp3:
    mean_critic_loss = pickle.load(fp3)

# PPO

with open('./Test Reward Plot/test_rewards_ppo', 'rb') as fp1:
    test_rewards_ppo     = pickle.load(fp1)
with open('./Loss Plot/mean_actor_loss_ppo', 'rb') as fp2:
    mean_actor_loss_ppo  = pickle.load(fp2)
with open('./Loss Plot/mean_critic_loss_ppo', 'rb') as fp3:
    mean_critic_loss_ppo = pickle.load(fp3)

# HPPO Beta No Clipped Reward Normalized
with open('./Test Reward Plot/test_rewards_beta', 'rb') as fp1:
    test_rewards_hbppo     = pickle.load(fp1)
with open('./Loss Plot/mean_actor_loss_beta', 'rb') as fp2:
    mean_actor_loss_hbppo  = pickle.load(fp2)
with open('./Loss Plot/mean_critic_loss_beta', 'rb') as fp3:
    mean_critic_loss_hbppo = pickle.load(fp3)
    
# Model Based Performance
with open ('./Test Reward Plot/mb_test_rewards', 'rb') as fp1:
    mb_test_rewards = pickle.load(fp1)
    
# Goal distance
with open('./Goal Plot/mb_goal3', 'rb') as fp1:
    mb_goal = pickle.load(fp1)

with open('./Goal Plot/sf_goal3', 'rb') as fp2:
    sf_goal = pickle.load(fp2)
```

# Generate Matlab File

```python
import numpy, scipy.io
scipy.io.savemat('test_rewards_ppo_pendulum.mat', mdict={'returns': test_rewards_ppo})
scipy.io.savemat('test_rewards_hppo_pendulum.mat', mdict={'returns': test_rewards})
scipy.io.savemat('test_rewards_hbppo_pendulum.mat', mdict={'returns': test_rewards_hbppo})
scipy.io.savemat('test_rewards_mbppo_pendulum.mat', mdict={'returns': mb_test_rewards})

scipy.io.savemat('mb_goal_pendulum.mat', mdict={'distance': mb_goal})
scipy.io.savemat('sf_goal_pendulum.mat', mdict={'distance': sf_goal})
```

# Plot Smoothed Curves

```python
 # Plot the episode reward over time
smoothing_window = 350

fig1 = plt.figure(figsize=(10,5))
test_rewards_smoothed = pd.Series(test_rewards).rolling(smoothing_window, min_periods=smoothing_window).mean()
plt.plot(test_rewards_smoothed)
plt.xlabel("Episode")
plt.ylabel("Episode Reward (Smoothed)")
plt.title("Episode Reward over Time (Smoothed over window size {}) - HPPO-ClippedNormal".format(smoothing_window))

plt.show(fig1)

fig2 = plt.figure(figsize=(10,5))
mean_actor_loss_smoothed = pd.Series(mean_actor_loss[:400]).rolling(smoothing_window, min_periods=smoothing_window).mean()
plt.plot(mean_actor_loss_smoothed)
plt.xlabel("Episode")
plt.ylabel("Mean Actor Loss (Smoothed)")
plt.title("Mean Actor Loss over Time (Smoothed over window size {}) - HPPO-ClippedNormal".format(smoothing_window))

plt.show(fig2)

fig3 = plt.figure(figsize=(10,5))
mean_critic_loss_smoothed = pd.Series(mean_critic_loss[:400]).rolling(smoothing_window, min_periods=smoothing_window).mean()
plt.plot(mean_critic_loss_smoothed)
plt.xlabel("Episode")
plt.ylabel("Mean Critic Loss (Smoothed)")
plt.title("Mean Critic Loss over Time (Smoothed over window size {}) - HPPO-ClippedNormal".format(smoothing_window))

plt.show(fig3)
```

```python
fig4 = plt.figure(figsize=(10,5))
test_rewards_ppo_smoothed = pd.Series(test_rewards_ppo).rolling(smoothing_window, min_periods=smoothing_window).mean()
plt.plot(test_rewards_ppo_smoothed)
plt.xlabel("Episode")
plt.ylabel("Episode Reward (Smoothed)")
plt.title("Episode Reward over Time (Smoothed over window size {}) - PPO-Normal".format(smoothing_window))

plt.show(fig4)

fig5 = plt.figure(figsize=(10,5))
mean_actor_loss_ppo_smoothed = pd.Series(mean_actor_loss_ppo).rolling(smoothing_window, min_periods=smoothing_window).mean()
plt.plot(mean_actor_loss_ppo_smoothed)
plt.xlabel("Episode")
plt.ylabel("Mean Actor Loss (Smoothed)")
plt.title("Mean Actor Loss over Time (Smoothed over window size {}) - PPO-Normal".format(smoothing_window))

plt.show(fig5)

fig6 = plt.figure(figsize=(10,5))
mean_critic_loss_ppo_smoothed = pd.Series(mean_critic_loss_ppo).rolling(smoothing_window, min_periods=smoothing_window).mean()
plt.plot(mean_critic_loss_ppo_smoothed)
plt.xlabel("Episode")
plt.ylabel("Mean Critic Loss (Smoothed)")
plt.title("Mean Critic Loss over Time (Smoothed over window size {}) - PPO-Normal".format(smoothing_window))

plt.show(fig6)
```

```python
fig7 = plt.figure(figsize=(10,5))
test_rewards_hbppo_smoothed = pd.Series(test_rewards_hbppo).rolling(smoothing_window, min_periods=smoothing_window).mean()
plt.plot(test_rewards_hbppo_smoothed)
plt.xlabel("Episode")
plt.ylabel("Episode Reward (Smoothed)")
plt.title("Episode Reward over Time (Smoothed over window size {}) - HPPO-Beta".format(smoothing_window))

plt.show(fig7)

fig8 = plt.figure(figsize=(10,5))
mean_actor_loss_hbppo_smoothed = pd.Series(mean_actor_loss_hbppo[:25000]).rolling(smoothing_window, min_periods=smoothing_window).mean()
plt.plot(mean_actor_loss_hbppo_smoothed)
plt.xlabel("Episode")
plt.ylabel("Mean Actor Loss (Smoothed)")
plt.title("Mean Actor Loss over Time (Smoothed over window size {}) - HPPO-Beta".format(smoothing_window))

plt.show(fig8)

fig9 = plt.figure(figsize=(10,5))
mean_critic_loss_hbppo_smoothed = pd.Series(mean_critic_loss_hbppo[:25000]).rolling(smoothing_window, min_periods=smoothing_window).mean()
plt.plot(mean_critic_loss_hbppo_smoothed)
plt.xlabel("Episode")
plt.ylabel("Mean Critic Loss (Smoothed)")
plt.title("Mean Critic Loss over Time (Smoothed over window size {}) - HPPO-Beta".format(smoothing_window))

plt.show(fig9)
```

```python
# Model Based
fig10 = plt.figure(figsize=(10,5))
mb_test_rewards_smoothed = pd.Series( mb_test_rewards).rolling(smoothing_window, min_periods=smoothing_window).mean()
plt.plot(mb_test_rewards_smoothed)
plt.xlabel("Episode")
plt.ylabel("Episode Reward (Smoothed)")
plt.title("Episode Reward over Time (Smoothed over window size {}) - Model Based".format(smoothing_window))

plt.show(fig10)
```
